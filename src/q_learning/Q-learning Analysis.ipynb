```python
import json
import os 
import sys
```

```python
data_path = os.path.abspath("__file__" + "../../../../data")
log_path = os.path.join(data_path, "log")

# 100 / 1000
update = 100

# 0.001 / 0.0001 / 0.00001
lr = 0.0001

log_name = f"_update_{update}_{lr}_penyes_preeliminated_norno-None_GNNGL_GlobalDecoder.log"
log_path = os.path.join(log_path, log_name)
```

```python
def process(log_file):
    loss = []
    success = []

    for line in log_file:
        
        if "loss:" in line:
            loss.append(float(line.split()[-1]))
        if "success count:" in line:
            success.append(int(line.split()[-1]))
        
    return loss, success
```

```python
with open (log_path, 'r') as log_file:
    loss, success = process(log_file)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(len(loss)), loss)
plt.title(f"DQN-loss-lr{lr}-update{update}")
```

```python
plt.plot(range(len(success)), success)
plt.title(f"DQN-success-lr{lr}-update{update}")
plt.ylim(0,3)
```
