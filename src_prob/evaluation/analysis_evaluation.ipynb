```python
import os
import json 
import re

data_dir = os.path.abspath(os.path.join(os.path.abspath("__file__"), "../data"))
model_dir = os.path.abspath(os.path.join(data_dir, "oak_model/model"))
log_dir = os.path.abspath(os.path.join(data_dir, "log"))

log_file_name = "search_500_result_hard_constraint.log"
log_path = os.path.join(log_dir, log_file_name)
```

```python
from copy import deepcopy

result = {}
prog_id = -1
res_ct = 0

with open(log_path, 'r') as log_file:
    for line in log_file: 
        
        res = re.search("root - INFO - ([0-9]*)\n", line)
        if not type(res) == type(None):
            prog_id = res.group(1)
            prog_info = {}
            result[prog_id] = {}
            res_ct = 0
        
        res = re.search("root - INFO -  prob: tensor((.*)), clause: (.*), suc: (.*)\n", line)
            
        if not type(res) == type(None):
            prob = res.group(2)
            if "grad_fn" in prob:
                prob = prob.split("]")[0]
                prob = prob[2:]
            else:
                prob = prob[2: -2]
            prog_info["prob"] = float(prob)
            prog_info["clauses"] = res.group(3)
            prog_info["suc"] = res.group(4)
            result[prog_id][res_ct] = deepcopy(prog_info)
            res_ct += 1
         
```

```python
suc_res = {}
for prog_id, prog_res_dict in result.items():
    suc_ct = 0
    beam_ct = 0
    for search_id, res in prog_res_dict.items():
        beam_ct += 1
        if res['suc'] == 'True':
            suc_ct += 1
    if beam_ct == 0:
        suc_res[prog_id] = 0
    else:
#         print (suc_ct)
#         print (beam_ct)
        suc_res[prog_id] = suc_ct / beam_ct
        
```

```python
fails = []

for prog_id, value in suc_res.items():
    if value == 0:
        fails.append(prog_id)
        
print(fails)
```

```python
fail_union = []
for prog in fails:
    if int(prog) in fail_progs:
        fail_union.append(prog)
print(fail_union)
```

```python
bin_num = 10
bin_count = [0] * bin_num
for prog_id, suc_val in suc_res.items():
    for i in range(bin_num):
        if not (i == bin_num -1) and suc_val >= i/bin_num and suc_val < (i+1)/bin_num:
            bin_count[i] += 1
            break
        if (i == bin_num - 1) and suc_val >= i/bin_num and suc_val <= (i+1)/bin_num:
            bin_count[i] += 1
```
